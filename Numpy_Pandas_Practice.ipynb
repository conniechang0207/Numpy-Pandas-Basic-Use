### 2) Re-implement the following logic in a single line of code with list comprehension

```python
# original code
def foo(n):
    return n*(n-1)/2

lst = [10, 20, 30, 40, 50]

nlst = []
for l in lst:
    nlst.append(foo(l))

print (nlst)
```

```python
# your single line of code using list comprehension here
nlst1 = list(n*(n-1)/2 for n in lst)
print(nlst1)
```

### 3) Re-implement the following logic in a single line of code with map and lambda

```python
# original code
def foo(n):
    return n*(n-1)/2

lst = [10, 20, 30, 40, 50]

nlst = []
for l in lst:
    nlst.append(foo(l))

print (nlst)
```

```python
# your single line of code using map and lambda here
nlst2 = list(map((lambda x: x*(x-1)/2),lst))
print(nlst2)
```

### 4) Re-implement the following logic in a single line of code with filter and lambda

```python
# original code
def foo(n):
    if n>25:
        return True
    else:
        return False

lst = [10, 20, 30, 40, 50]

nlst = []
for l in lst:
    if foo(l) == True:
        nlst.append(l)

print (nlst)
```

```python
# your single line of code using filter and lambda here
nlst3 = list(filter((lambda x: x>25),lst))
print(nlst3)
```
